# Probabilidade - Tarefa

```python
import numpy as np
from scipy import stats
from scipy.stats import norm, t
```

1. Uma máquina embala ração para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatório, com distribuição Normal, com média 15kg, mas com desvio padrão 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produção tenha mais que 15,2 kg.

```python
# Média e desvio padrão da distribuição normal
media = 15.0  # Média de 15 kg
desvio_padrao = 0.1  # Desvio padrão de 0.1 kg

# Ponto para o qual queremos calcular a probabilidade
peso_limite = 15.2

# Calcular a probabilidade de que um pacote pese mais que 15.2 kg
probabilidade = 1 - norm.cdf(peso_limite, loc=media, scale=desvio_padrao)

print(f'A probabilidade de que um pacote pese mais que 15,2 kg é {probabilidade:.5f}')

```

2. Da máquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Média e desvio padrão da distribuição normal
media = 15.0  # Média de 15 kg
desvio_padrao = 0.1  # Desvio padrão de 0.1 kg

# Percentil de ordem 95% para a distribuição normal
percentil_95 = norm.ppf(0.95, loc=media, scale=desvio_padrao)

print(f'O percentil de ordem 95% para o peso dos pacotes é {percentil_95:.5f} kg')
```

3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na célula abaixo:

```python
Y=np.linspace(-6, 6, 200)
# Seu código começa aqui

# Definição do objeto Y com valores para calcular as probabilidades
Y = np.array([0.5, 1.0, 1.5, 2.0])

# Calcular as probabilidades para Y_1 (Normal Padrão)
probabilidades_Y1 = norm.cdf(Y, loc=0, scale=1)  # Média = 0, desvio padrão = 1

# Calcular as probabilidades para Y_2 (t-Student com 5 graus de liberdade)
graus_de_liberdade = 5
probabilidades_Y2 = t.cdf(Y, df=graus_de_liberdade)

# Exibir os resultados
print(f'Probabilidades de Y_1 ser menor que cada valor de Y:')
for y, p in zip(Y, probabilidades_Y1):
    print(f'P(Y_1 < {y}) = {p:.5f}')

print(f'\nProbabilidades de Y_2 ser menor que cada valor de Y:')
for y, p in zip(Y, probabilidades_Y2):
    print(f'P(Y_2 < {y}) = {p:.5f}')
```

4. Faça uma visualização gráfica dos dados do item anterior.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t

# Definição do array Y com valores para calcular as probabilidades
Y = np.array([0.5, 1.0, 1.5, 2.0])

# Calcular as probabilidades para Y_1 (Normal Padrão)
probabilidades_Y1 = norm.cdf(Y, loc=0, scale=1)  # Média = 0, desvio padrão = 1

# Calcular as probabilidades para Y_2 (t-Student com 5 graus de liberdade)
graus_de_liberdade = 5
probabilidades_Y2 = t.cdf(Y, df=graus_de_liberdade)

# Plotar o gráfico das funções de distribuição acumulada (CDF)
plt.figure(figsize=(10, 6))

# Linha para Normal Padrão
plt.plot(Y, probabilidades_Y1, label='Y_1 (Normal Padrão)', marker='o')

# Linha para t-Student com 5 graus de liberdade
plt.plot(Y, probabilidades_Y2, label='Y_2 (t-Student, 5 df)', marker='o')

# Configurações do gráfico
plt.title('Probabilidades acumuladas de Y_1 e Y_2')
plt.xlabel('Valores de Y')
plt.ylabel('Probabilidade acumulada')
plt.legend()
plt.grid(True)

# Mostrar o gráfico
plt.show()
```

5. 3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
Y=np.linspace(0, 1, 20)
# Seu código começa aqui

# Definição do array de probabilidades P
P = np.array([0.025, 0.5, 0.975])

# Calcular os percentis para Y_1 (Normal Padrão)
percentis_Y1 = norm.ppf(P, loc=0, scale=1)  # Média = 0, desvio padrão = 1

# Calcular os percentis para Y_2 (t-Student com 5 graus de liberdade)
graus_de_liberdade = 5
percentis_Y2 = t.ppf(P, df=graus_de_liberdade)

# Exibir os resultados
print(f"Percentis para Y_1 (Normal Padrão): {percentis_Y1}")
print(f"Percentis para Y_2 (t-Student, 5 df): {percentis_Y2}")
```
